```
# Reg. No: 21BDS0244    Name: Amitanshu Panigrahi
# Dataset: https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bodyPerformance.csv

# MODULE 1

# importing necessary libraries and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# loading the dataset
data_url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bodyPerformance.csv"
df = pd.read_csv(data_url)

# displaying the first five rows of the dataset
print("Head of the Dataset:")
print(df.head())

# Dimensions of the dataset
print("\nDataset Dimensions:")
print(df.shape)

# Summary of the Dataset
print("\nSummary:")
print(df.describe())

# missing values
print("\nMissing values:")
print(df.isnull().sum())

# Data types present in given dataset
print("\nData types:")
print(df.dtypes)

```

Module 2

```
# MODULE 2

# Data Transformation: Merging, Concatenation, Joins

# adding a new column based on other attributes (eg. bmi)
df['BMI'] = df['weight_kg'] / (df['height_cm'] / 100) ** 2
print("\nBMI column created:")
print(df[['height_cm', 'weight_kg', 'BMI']].head())

# splitting dataset into two
dp1 = df.iloc[:len(df)//2]
dp2 = df.iloc[len(df)//2:]

# merging based on a common column (age)
mdata = pd.merge(dp1, dp2, on='age', how='inner', suffixes=('_dp1', '_dp2'))
print("\nmerged dataframe (inner join):")
print(mdata.head())

# concatenation along rows
cdatarow = pd.concat([dp1, dp2], axis=0)
print("\nconcatenated dataframe (rows):")
print(cdatarow.head())

# concatenation along columns
cdatacol = pd.concat([dp1.reset_index(drop=True), dp2.reset_index(drop=True)], axis=1)
print("\nconcatenated dataframe (columns):")
print(cdatacol.head())

# outer join
outer_data = pd.merge(dp1, dp2, on='age', how='outer', suffixes=('_dp1', '_dp2'))
print("\nouter join dataframe:")
print(outer_data.head())

# left join
left_data = pd.merge(dp1, dp2, on='age', how='left', suffixes=('_dp1', '_dp2'))
print("\nleft join dataframe:")
print(left_data.head())

# right join
right_data = pd.merge(dp1, dp2, on='age', how='right', suffixes=('_dp1', '_dp2'))
print("\nright join dataframe:")
print(right_data.head())

# stack and unstack operations
stacked = df.stack()
unstacked = stacked.unstack()
print("\nstacked dataframe:")
print(stacked.head())
print("\nunstacked dataframe:")
print(unstacked.head())

# checking for duplicates
duplicates = df[df.duplicated()]
print("\nduplicated rows:")
print(duplicates)

# drop duplicates
df_nodup = df.drop_duplicates()
print("\ndataframe after dropping duplicates:")
print(df_nodup.shape)


# handling missing values
# forward fill
df_ffill = df.fillna(method='ffill')
print("\nmissing values handled using forward fill:")
print(df_ffill.isnull().sum())

# backward fill
df_bfill = df.fillna(method='bfill')
print("\nmissing values handled using backward fill:")
print(df_bfill.isnull().sum())

# replace with mean
df_mean = df.fillna(df.mean(numeric_only=True))
print("\nmissing values handled by replacing with mean:")
print(df_mean.isnull().sum())

# replace with cumulative sum only for numeric columns
numeric_columns = df.select_dtypes(include=np.number).columns
df_cumsum = df.copy()  # create a copy to avoid modifying the original dataframe
df_cumsum[numeric_columns] = df_cumsum[numeric_columns].fillna(df_cumsum[numeric_columns].cumsum())
print("\nmissing values in numeric columns handled using cumulative sum:")
print(df_cumsum.isnull().sum())

# drop rows with missing values
df_droprows = df.dropna()
print("\ndataframe after dropping rows with missing values:")
print(df_droprows.isnull().sum())

# drop columns with missing values
df_dropcols = df.dropna(axis=1)
print("\ndataframe after dropping columns with missing values:")
print(df_dropcols.isnull().sum())
```

```
bmi_bins = [0, 18.5, 24.9, 29.9, 100]
bmi_labels = ['underweight', 'normal weight', 'overweight', 'obese']
df['BMI_category'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels, right=False)
print("\nBMI categories created:")
print(df[['BMI', 'BMI_category']].head())

plt.figure(figsize=(6, 4))
sns.countplot(x=df['BMI_category'], palette='magma')
plt.title('BMI category distribution')
plt.xlabel('BMI category')
plt.ylabel('count')
plt.show()
```

```
# outlier detection and filtering (eg. for 'BMI')
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['BMI'], palette='Set2')
plt.title('Boxplot for BMI (Outlier Detection)')
plt.xlabel('BMI')
plt.show()

# removing outliers based on IQR
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_df = df[(df['BMI'] >= lower_bound) & (df['BMI'] <= upper_bound)]
print("\nDataFrame after filtering outliers in BMI:")
print(filtered_df.shape)
```

```
# MODULE 3

# Univariate analysis

# loading the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bodyPerformance.csv"
df = pd.read_csv(url)

# checking data types of each column
print("Data Types of Each Column:")
print(df.dtypes)

# calculating descriptive statistics of the entire dataset
print("\ndescriptive statistics of the dataset:")
print(df.describe(include='all'))

# finding the number of non-numeric values in each column
non_num_count = df.apply(lambda col: col.apply(lambda x: not str(x).replace('.', '', 1).isdigit()).sum())
print("\nnumber of non-numeric values in each column:")
print(non_num_count)

# listing non-numeric values in each column
non_num_vals = {
    col: df[col][df[col].apply(lambda x: not str(x).replace('.', '', 1).isdigit())].unique()
    for col in df.columns
}
print("\nlist of non-numeric values in each column:")
print(non_num_vals)

# converting non-numeric values to nan and filling missing values with the column mean
df_cleaned = df.apply(pd.to_numeric, errors='coerce')
df_cleaned.fillna(df_cleaned.mean(), inplace=True)

# computing measures of central tendency for 'height_cm'
col = 'height_cm'
mean = df_cleaned[col].mean()
median = df_cleaned[col].median()
mode = df_cleaned[col].mode()[0]

print(f"\nmeasures of central tendency for {col}:")
print(f"mean: {mean}, median: {median}, mode: {mode}")

# bar chart of the first 30 largest values in 'height_cm' column
top_30 = df_cleaned.nlargest(30, col)
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_30[col])), top_30[col])
plt.xlabel('Index')
plt.ylabel(col)
plt.title(f'bar chart of top 30 largest values in {col}')
plt.show()

# distribution plot for 'height_cm' columb
g = sns.FacetGrid(df).map(sns.distplot,"height_cm").add_legend()
g.set_axis_labels(col, 'Density')
g.fig.suptitle(f'distribution plot of {col}', y=1.02)
plt.show()

# boxplot for 'height_cm' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_cleaned[col])
plt.title(f'Boxplot of {col}')
plt.xlabel(col)
plt.show()
```

```
# Bivariate analysis

# scatterplot between 'height_cm' and 'weight_kg'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='height_cm', y='weight_kg', data=df_cleaned)
plt.show()

# relationship between two features (positive correlation)
positive_corr = np.corrcoef(df_cleaned["height_cm"], df_cleaned["weight_kg"])
print("Correlation (positive) between height_cm and weight_kg:")
print(positive_corr)

# scatterplot between 'gripForce' and 'sit and bend forward_cm'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gripForce', y='sit and bend forward_cm', data=df_cleaned)
plt.show()

# relationship between two features (negative correlation)
negative_corr = np.corrcoef(df_cleaned["gripForce"], df_cleaned["sit and bend forward_cm"])
print("Correlation (negative) between gripForce and sit and bend forward_cm:")
print(negative_corr)


# boxplot between 'height_cm' and 'weight_kg' grouped by class
sns.boxplot(x="height_cm", y="class",data=df)
```

```
# Multivariate analysis

# selected columns for pair plot
selected_columns = ['height_cm', 'weight_kg', 'gripForce', 'sit and bend forward_cm', 'body fat_%']
df_subset = df_cleaned[selected_columns]

# pair plot with regression plot type, custom color, and diagonal KDE
sns.pairplot(
    df_subset,
    kind="reg",  # regression for scatterplots
    diag_kind="kde",  # KDE for diagonal
    palette="coolwarm",  # custom color palette
)
plt.suptitle("Pair Plot with Regression (Selected Columns)", y=1.02)
plt.show()

# calculating correlation coefficient
from scipy import stats

corr = stats.pearsonr(df["height_cm"], df["weight_kg"])
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])

# finding correlation using pearson method
correlation_matrix = df_cleaned.corr(method='pearson')

# visualizing this correlation analysis using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Heatmap of Numerical Records")
plt.show()
```

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Load dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bodyPerformance.csv"
df = pd.read_csv(url)

# Preview data
print(df.head())
```

```
# 1D Statistical Data Analysis

# Central Tendency
# Mean
mean_val = df['height_cm'].mean()
mean_trimmed = stats.trim_mean(df['height_cm'], 0.1)
print("Mean:", mean_val)
print("Trimmed Mean:", mean_trimmed)

# Median, Quantile, Decile, Percentile
median_val = df['height_cm'].median()
quantiles = df['height_cm'].quantile([0.25, 0.5, 0.75])
deciles = df['height_cm'].quantile(np.arange(0.1, 1, 0.1))
percentiles = np.percentile(df['height_cm'], [25, 50, 75])

print("Median:", median_val)
print("Quantiles:\n", quantiles)
print("Deciles:\n", deciles)
print("Percentiles:\n", percentiles)
```

```
# Measure of Dispersion
# Range, IQR, Interdecile range
range_val = df['height_cm'].max() - df['height_cm'].min()
iqr_val = stats.iqr(df['height_cm'])
interdecile_range = np.percentile(df['height_cm'], 90) - np.percentile(df['height_cm'], 10)

# Mean deviation, Standard deviation
mean_dev = np.mean(np.abs(df['height_cm'] - df['height_cm'].mean()))
std_dev = df['height_cm'].std()

# Skewness and Kurtosis
skew = df['height_cm'].skew()
kurt = df['height_cm'].kurt()

print("Range:", range_val)
print("IQR:", iqr_val)
print("Interdecile Range:", interdecile_range)
print("Mean Deviation:", mean_dev)
print("Standard Deviation:", std_dev)
print("Skewness:", skew)
print("Kurtosis:", kurt)
```

```
# Frequency Distribution
sns.histplot(df['height_cm'], kde=True)
plt.title('Frequency Distribution of Height')
plt.show()

# Relative Frequency
freq = df['height_cm'].value_counts(normalize=True)
print(freq)

# Cumulative Frequency
cumulative_freq = freq.cumsum()
print(cumulative_freq)
```

```
# Pie Plot
df['gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()

# Stacked Bar Plot
pd.crosstab(df['class'], df['gender']).plot(kind='bar', stacked=True)
plt.title('Class vs Gender')
plt.show()
```

```
# 2D Statistical Data Analysis
# 2-Way Contingency Table

# Categorical-Categorical
print(pd.crosstab(df['class'], df['gender']))

# Numerical-Numerical
print(df[['height_cm', 'weight_kg']].corr())

# Categorical-Numerical
print(df.groupby('gender')['height_cm'].mean())
```

```
#3-Way Contingency Table

three_way = pd.crosstab(index=[df['class'], df['gender']], columns=df['age'])
print(three_way)
```

```
# Row Profile, Column Profile, Relative Frequency, Chi-Square

row_profile = three_way.div(three_way.sum(axis=1), axis=0)
col_profile = three_way.div(three_way.sum(axis=0), axis=1)
relative_freq = three_way / three_way.sum().sum()

from scipy.stats import chi2_contingency
chi2, p, _, _ = chi2_contingency(three_way)

print("Row Profile:\n", row_profile)
print("Column Profile:\n", col_profile)
print("Relative Frequency:\n", relative_freq)
print("Chi-Square Value:", chi2)
print("P-Value:", p)
```

```
# 3d Bar plot
# Grouping by gender and class
freq = df.groupby(['gender', 'class']).size().unstack(fill_value=0)

xpos, ypos = np.meshgrid(np.arange(freq.shape[0]), np.arange(freq.shape[1]))
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)

dx = dy = 0.5
dz = freq.values.flatten()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)

ax.set_xlabel('Gender')
ax.set_ylabel('Class')
ax.set_zlabel('Frequency')
plt.title('3D Bar Plot')
plt.show()
```
